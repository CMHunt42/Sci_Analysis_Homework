5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np

class Particle(object):
    def __init__(self, mass, x, y, z, vx, vy, vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
    
    def K_energy(self):
        print(0.5 * self.mass * ((self.vx**2) + (self.vy**2) + (self.vz**2)))
        
    def distance(self, p2):
        print(np.sqrt((self.x - p2.x)**2 + (self.y - p2.y)**2 + (self.z - p2.z)**2))

    def newPos(self, dt):
        self.x += (self.vx * dt)
        self.y += (self.vy * dt)
        self.z += (self.vz * dt)


class ChargedParticle(Particle):

    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.charge = charge
```
